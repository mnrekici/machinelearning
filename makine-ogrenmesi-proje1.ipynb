```python
from sklearn.linear_model import LinearRegression
import pandas as pd
df=pd.read_csv("/kaggle/input/student-marks-dataset/Student_Marks.csv")
df.head(3)
```

```python
y=df[["Marks"]]
x=df[["number_courses","time_study"]]
```

```python
df.info()
```

```python
l=LinearRegression()
model=l.fit(x,y) #fit öğrenmek demek, öğreteceğiz. x ve y elimizdeki veriler
```

```python
model.predict([[4,4]]) #4.sınıf, 4 saat çalışıyor, kaç alır tahmin et demiş olduk modele
```

```python
df['Marks'].max() #alınacak maximum puan
```

```python
model.score(x,y) # modelin başarısı
```
