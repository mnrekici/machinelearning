```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
df=pd.read_csv("/kaggle/input/audi-a1-listings/Audi_A1_listings.csv")
df.head(3)

df=df.drop(columns=["index","href","MileageRank","PriceRank","PPYRank","Score"])
df.head(3)
```

```python
df.info
```

```python
df.head(3)
```

```python
df.columns=["yil","kasa","mil","motor","ps","vites","yakit","sahip","fiyat","ppy"]
df.head(3)
```

```python
df["motor"] = df["motor"].astype(str)
df["motor"] = df["motor"].str.replace("L", "")  # "L" harfini kaldır
df["motor"] = pd.to_numeric(df["motor"])
```

```python
df.head(3) 
```

```python
df=pd.get_dummies(df,columns=["kasa","vites","yakit"],drop_first=True, dtype=int)
df.head(3)
```

```python
y=df[["fiyat"]]
x=df.drop("fiyat",axis=1) #axis=1 sütunlardan fiyatı sil demek 
```

```python
lm = LinearRegression()
model=lm.fit(x,y)
model.predict([[2017,30000,1.6,110,1,2600,0,1]])
```
