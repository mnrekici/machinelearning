```python
from prophet import Prophet
import yfinance as yf
import pandas as pd
```

```python
df=yf.download('BTC-USD',"2015-01-01","2024-04-26")
```

```python
df=df[['Close']]
```

```python
df=df.reset_index()
```

```python
df.head(3)
```

```python
df.columns=['ds','y']
```

```python
df.head(3)
```

```python
model=Prophet()
model.fit(df)
```

```python
gelecek=model.max_future_dataframe()
tahmin=model.predict(gelecek)
```

```python
model.plot(tahmin);
```

```python
model.plot_components(tahmin);
```
